```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Lenovo\Desktop\lead_pre.csv")
df
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace = True)
```

```python
print(df.describe())
```

```python
# Countplot of the 'status' column
sns.countplot(x='status', data=df)
plt.title('Status')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='lost_reason', data=df, order=df['lost_reason'].value_counts().index)
plt.title('Lost Reason')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['budget'], kde=True, bins=20)
plt.title('Budget')
plt.xticks(rotation=45)
plt.gca().invert_xaxis()  # Invert the x-axis to sort bins in descending order
plt.show()
```

```python
# Scatter plot of 'budget' vs. 'duration'
plt.figure(figsize=(10, 5))
sns.scatterplot(x='budget', y='duration', data=df)
plt.title('Budget vs. Duration')
plt.show()
```

```python
# Distribution Analysis
df.hist(figsize=(10, 8), bins=20)
plt.tight_layout()
plt.show()
```

```python
# Outlier Detection
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['budget', 'duration']])
plt.title('Outlier Detection')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['status'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Lead Status')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['lost_reason'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Lost Reason')
plt.ylabel('')
plt.show()
```
